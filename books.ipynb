```python
import requests as req
from bs4 import BeautifulSoup as bs
from pprint import pprint
import json
import re
```

```python
url='https://www.gutenberg.org/browse/languages/zh'
res= req.get(url)
res.encoding = 'utf-8'
#print(res.text)
soup = bs(res.text, "lxml")
```

```python
list_book = []
list_bookUrl = []
regex01 = r'/\d+' # /+數字 比對抓到書的ID
regex02 = r'\r'
#  and r'\s'
for a in soup.select("li.pgdbetext>a"):
    if '\u4e00'<= a.text[0] <='\u9fa5': # 判斷書名的第一個字為中文
        fileName = re.search(regex01,a['href']) # 取屬型為href的值
        bookName = re.sub(regex02,'-',a.text)
        # print(fileName)
        print(bookName)
        # book_id = a['href'].split('/')[-1] #利用切割抓到就可以抓到id
        # print(book_id)
        #list_bookUrl.append('https://www.gutenberg.org/files/'+ (book_id+"/")*2+'-0.txt')
        list_bookUrl.append('https://www.gutenberg.org/files'+(fileName.group())*2+'-0.txt') # fileName.group()抓出ID
        list_book.append(bookName) # 把a的文字加進書名
        
       # print(list_book)
list_book
```

```python
booksurl_text = []
regex03 = r'[^a-zA-Z\s\d*#,.\-:\]\[<>\/\!\"\(\)\@\$\'\%]+' # 抓出這些符號都不要
count = 0
for link in list_bookUrl[0:250]:
    res = req.get(link)
    res.encoding = 'UTF-8'
    soup = bs(res.text,'lxml')
    text = re.findall(regex03, soup.text) #取出沒有英文和符號的內文
    textj = '\n'.join(text) # 根據正規表達式裡不要的東西 來分隔
    path = f'books/{list_book[count]}.txt' # 路徑要放到哪
    f = open(path, 'w', encoding = 'UTF-8')
    f.write(f'書名：{list_book[count]}'+'\n'+textj)
    f.close()
    count += 1
```
